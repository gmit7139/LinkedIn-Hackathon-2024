```python
# Top 5 Most Popular Courses Taken by Users

# Analysis functions
def analyze_user_data(user_df):
    required_columns = ['name', 'length']
    check_columns(user_df, required_columns)

    # Most popular courses taken by users
    courses = user_df['name'].dropna().tolist()
    most_common_courses = Counter(courses).most_common(5)

    # Calculate the average length
    lengths = user_df['length'].dropna().apply(lambda x: x['value'] if x['unit'] == 'hours' else x['value'] / 60)
    average_length = lengths.mean()

    return most_common_courses, average_length

# Perform analysis
most_common_courses, average_length = analyze_user_data(user_df)

# Print results
print("Top 5 Most Popular Courses Taken by Users:")
for course, count in most_common_courses:
    print(f"{course}: {count} times")

print(f"\nAverage Course Length: {average_length:.2f} hours")

# Extracting course names and their counts
courses, counts = zip(*most_common_courses)

# Creating the bar graph
plt.figure(figsize=(6, 6))
plt.bar(courses, counts, color='#0072b1')
plt.xlabel('Courses')
plt.ylabel('Number of Users')
plt.title('Top 5 Most Popular Courses Taken by Users')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Analysis
def analyze_user_data(user_df):
    required_columns = ['level']
    check_columns(user_df, required_columns)

    # Calculate the count and percentage of each difficulty level
    levels = user_df['level'].dropna().tolist()
    level_counts = Counter(levels)
    total_levels = sum(level_counts.values())
    level_percentages = {level: (count / total_levels) * 100 for level, count in level_counts.items()}

    return level_counts, level_percentages

# Perform analysis
level_counts, level_percentages = analyze_user_data(user_df)

# Print results
print("\nCount and Percentage of Each Difficulty Level:")
for level in level_counts:
    print(f"{level}: {level_counts[level]} ({level_percentages[level]:.2f}%)")

# Create a bar graph for difficulty levels
levels, percentages = zip(*level_percentages.items())
counts = [level_counts[level] for level in levels]

fig, ax1 = plt.subplots(figsize=(6, 6))

# Bar graph for percentages
color = '#0072b1'
ax1.set_xlabel('Difficulty Level')
ax1.set_ylabel('Percentage of Courses', color='black')
ax1.bar(levels, percentages, color=color, alpha=0.7)
ax1.tick_params(axis='y', labelcolor='black')

plt.title('Count and Percentage of Courses by Difficulty Level')
fig.tight_layout()

# Display the plot
plt.show()
```
